```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(X_train_cv, y_train) #model is trained
```

```python
X_test_cv = v.transform(X_test.values) #count vectorized at the time of X_test transform only transform() is used not fit_transform()
```

```python
from sklearn.metrics import classification_report #it is used becouse ham and spam are imbalanced if balanced then used accuracy score
```

```python
y_pred = model.predict(X_test_cv)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb', MultinomialNB())
]) 
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
```
