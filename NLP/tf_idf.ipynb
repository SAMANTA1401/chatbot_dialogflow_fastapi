```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = [
    "Thor eating pizza, Loki is eating pizza, Ironman ate pizza already",
    "Apple is announcing new iphone tomorrow",
    "Tesla is announcing new model-3 tomorrow",
    "Google is announcing new pixel-6 tomorrow",
    "Microsoft is announcing new surface tomorrow",
    "Amazon is announcing new eco-dot tomorrow",
    "I am eating biryani and you are eating grapes"

]
```

```python
v = TfidfVectorizer()
transformed_output = v.fit_transform(corpus)
```

```python
print(v.vocabulary_) #with indexing
```

```python
dir(v)
```

```python
all_feature_names = v.get_feature_names_out()

print(all_feature_names) #all words in the vocabulary
```

```python
for word in all_feature_names:
    idfscore = v.vocabulary_.get(word)
    print(f'{word} {v.idf_[idfscore]}')  #idf score
    
```

```python
import pandas as pd
```

```python
df = pd.read_csv('ecommerceDataset.csv',names=['label','text'])
print(df.shape)
```

```python
df['label_num'] =  df.label.map({
    'Household':0,
    'Books':1,
    'Electronics':2,
    'Clothing & Accessories':3
})
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.text, 
    df.label_num, 
    test_size=0.2, 
    random_state=42, 
    stratify=df.label_num #the stratify=y parameter ensures that the same proportion of each class (0, 1, and 2) is maintained in both the training and testing sets
)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
clf = Pipeline([
    ('vectorizer_tfidf',TfidfVectorizer()),
    ('KNN',KNeighborsClassifier())
])

clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
# utility function for pre-processing the test
import spacy
```

```python
#load english
nlp = spacy.load('en_core_web_sm')
```

```python
def preprocess(text):
    '''remove stop word and lemmatize the text'''
    doc = nlp(text)
    filtered_token = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_token.append(token.lemma_)
    # return filtered_token
    return " ".join(filtered_token) # create a string from a list of word
```

```python
df['preprocessed_text'] = df['text'].apply(preprocess)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.preprocessed_text, 
    df.label_num, 
    test_size=0.2, 
    random_state=42, 
    stratify=df.label_num #the stratify=y parameter ensures that the same proportion of each class (0, 1, and 2) is maintained in both the training and testing sets
)
```

```python
clf = Pipeline([
    ('vectorizer_tfidf',TfidfVectorizer()),
    # ('KNN',KNeighborsClassifier())
    ('Random forest',RandomForestClassifier())
])

clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```
